```python
#Importing Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df=pd.read_csv('/content/fatalities_isr_pse_conflict_2000_to_2023.csv')
df.transpose()
```

```python
df.head()#Glancing the certain data's from the Dataset
```

```python
#Checking any null or Duplicated values
print("The Number of rows and Columns in the Dataset is: ",df.shape)
print("The Number of Duplicated values in Dataset is: ",df.duplicated().sum())
print("The Number of Missing values in Dataset is: ",sum(df.isna().sum()))
```

# ***Date_of_event and Date_of_death in Date-Time Format.***

```python
#Date_of_Event to Date-Time Format
df['Event_Date']=pd.to_datetime(df['date_of_event']).dt.date
df['Event_Day']=pd.to_datetime(df['date_of_event']).dt.day
df['Event_Month']=pd.to_datetime(df['date_of_event']).dt.month
df['Event_Year']=pd.to_datetime(df['date_of_event']).dt.year

#Date_of_Death to Date-Time Format
df['Death_Date']=pd.to_datetime(df['date_of_death']).dt.date
df['Death_Day']=pd.to_datetime(df['date_of_death']).dt.day
df['Death_Month']=pd.to_datetime(df['date_of_death']).dt.month
df['Death_Year']=pd.to_datetime(df['date_of_death']).dt.year

print("The date_of_event after Date-Time format \n")
print(df[['Event_Date','Event_Day','Event_Month','Event_Year']].head())
print('\n')
print("The date_of_death after Date-Time format \n")
print(df[['Death_Date','Death_Day','Death_Month','Death_Year']].head())
```
